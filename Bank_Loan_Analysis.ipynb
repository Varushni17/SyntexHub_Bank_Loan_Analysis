```python
# Create a fresh copy
loan = df[['loan_amnt','funded_amnt','term','int_rate','installment',
           'grade','sub_grade','emp_length','home_ownership',
           'annual_inc','verification_status','issue_d',
           'loan_status','purpose','addr_state','dti',
           'total_pymnt']].copy()

# Convert interest rate to numeric
loan['int_rate'] = loan['int_rate'].str.rstrip('%').astype(float)

# Convert issue date
loan['issue_d'] = pd.to_datetime(
    loan['issue_d'],
    format='%b-%y',
    errors='coerce'
)

# Fill missing employment length
loan['emp_length'] = loan['emp_length'].fillna('Unknown')

# Verify
print(loan['int_rate'].dtype)
print(loan['int_rate'].head())
```

```python
# ==============================
# BANK LOAN KPI SUMMARY
# ==============================

total_applications = len(loan)
total_funded = loan['funded_amnt'].sum()
total_received = loan['total_pymnt'].sum()
avg_interest = loan['int_rate'].mean()
avg_dti = loan['dti'].mean()

print("="*50)
print("           BANK LOAN KPI SUMMARY")
print("="*50)
print(f"Total Loan Applications : {total_applications:,}")
print(f"Total Funded Amount     : ${total_funded:,.2f}")
print(f"Total Amount Received   : ${total_received:,.2f}")
print(f"Average Interest Rate   : {avg_interest:.2f}%")
print(f"Average DTI             : {avg_dti:.2f}%")
```

```python
# Good Loan = Fully Paid or Current
good_loans = loan[loan['loan_status'].isin(['Fully Paid', 'Current'])]

# Bad Loan = Charged Off
bad_loans = loan[loan['loan_status'] == 'Charged Off']

good_percent = len(good_loans) / len(loan) * 100
bad_percent = len(bad_loans) / len(loan) * 100

print("="*50)
print("         GOOD vs BAD LOAN")
print("="*50)
print(f"Good Loans : {len(good_loans):,} ({good_percent:.2f}%)")
print(f"Bad Loans  : {len(bad_loans):,} ({bad_percent:.2f}%)")
```

```python
import matplotlib.pyplot as plt

status_counts = loan['loan_status'].value_counts()

plt.figure(figsize=(8,5))
status_counts.plot(kind='bar')

plt.title("Loan Status Distribution")
plt.xlabel("Loan Status")
plt.ylabel("Number of Loans")

plt.show()
```

```python
plt.figure(figsize=(8,5))
loan['grade'].value_counts().sort_index().plot(kind='bar')

plt.title("Loan Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Number of Loans")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10,5))
loan['purpose'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Loan Purposes")
plt.xlabel("Purpose")
plt.ylabel("Number of Loans")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10,5))
loan['addr_state'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 States by Loan Applications")
plt.xlabel("State")
plt.ylabel("Applications")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
monthly = loan.groupby(loan['issue_d'].dt.to_period('M')).size()

plt.figure(figsize=(12,5))
monthly.plot()

plt.title("Monthly Loan Applications")
plt.xlabel("Month")
plt.ylabel("Applications")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(loan['int_rate'], bins=20)

plt.title("Interest Rate Distribution")
plt.xlabel("Interest Rate (%)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(loan['annual_inc'], bins=30)

plt.title("Annual Income Distribution")
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.show()
```
